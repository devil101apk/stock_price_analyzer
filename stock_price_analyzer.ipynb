```python
import numpy as np
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
import seaborn 
```

```python
stock=["AAPL", "AMZN", "TSLA"]
stocks=yf.download(stock, start ="2023-01-01", end="2025-01-01")
data=stocks.loc[:,"Close"].copy()
```

```python
data.plot(figsize=(17,8), fontsize=18)
plt.style.use("seaborn")
plt.show()
```
